```python
def find_java_files(root_path):
    java_files = []

    for root, dirs, files in os.walk(root_path):
        for file in files:
            if file.endswith(".java"):
                java_files.append(os.path.join(root, file))

    return java_files
```

```python
def append_bullet_numbers(input_string):
    lines = input_string.split('\n')
    result = []

    # Counter for bullet numbers
    bullet_number = 1

    for line in lines:
        if line.endswith('.java'):
            result.append(f"{bullet_number}. {line}")
            bullet_number += 1
        else:
            result.append(line)

    return '\n'.join(result)

def file_wise_summary(java_files_list):
    final_agg_summary=''
    for file_name in java_files_list:
        with open(file_name, 'r') as file:
            code_snippet=file.read()    
        trunc_file=os.path.basename(file_name)
        messages = [] 
        prompt=f"""Given the following code give the purpose and structure of the code in 100 words, in the given format.
        Format:
        {trunc_file}
          - Purpose
          - Structure
        Here is the code:\n+{code_snippet}."""
        messages.append({'role':'user','content':prompt})
        response = openai.ChatCompletion.create(
            engine="neww",
            messages = messages,
            temperature=0,  # Adjust the temperature for creativity
        )
        result=response.choices[0].message['content']
        #print(prompt)
        print(result)
        final_agg_summary=final_agg_summary+'\n'+result+'\n'
    return final_agg_summary
```

```python
final_agg_summary=file_wise_summary(java_files)
#print(final_agg_summary)
output_string = 'Files Breakdown:\n'+append_bullet_numbers(final_agg_summary)
print(output_string)
```

```python
def summarizer(java_files_list):
    summaries=''
    for file_name in java_files_list:
        with open(file_name, 'r') as file:
            code_snippet=file.read()    
        trunc_file=os.path.basename(file_name)
        messages = [] 
        prompt=f"""Summarize what the following code from {trunc_file} code does in 100 words.
        Here is the code:\n+{code_snippet}."""
        messages.append({'role':'user','content':prompt})
        response = openai.ChatCompletion.create(
            engine="neww",
            messages = messages,
            temperature=0.2,  # Adjust the temperature for creativity
        )
        result=response.choices[0].message['content']
        summaries=summaries+'\n'+trunc_file+':\n'+result
        #print(result)
    return summaries
```

```python
summaries=summarizer(java_files)
print(summaries)
```

```python
def overview(summaries):
    messages = [] 
    prompt=f"""Give a 100 word Overview based on the following, but do not go in detail
    with respect to any single functionality.
    Output format:
    Overview:
    ---content---
    \nHere are the summaries:\n{summaries}
    """
    messages.append({'role':'user','content':prompt})
    response = openai.ChatCompletion.create(
        engine="neww",
        messages = messages,
        temperature=0.2,  # Adjust the temperature for creativity
    )
    result=response.choices[0].message['content']
    return result
```

```python
overview_section=overview(summaries)
overview_section
```

```python
def userflow(summaries):
    messages = [] 
    prompt=f"""Given below is a text containing various files and their descriptions. Based on these give one example
    of how this application might work as a whole. Do not explain individual files with example but explain as an overall example.
    Output format:
    Example:
    ---content---
    \nHere is the data:\n{summaries}
    """
    messages.append({'role':'user','content':prompt})
    response = openai.ChatCompletion.create(
        engine="neww",
        messages = messages,
        temperature=0.2,  # Adjust the temperature for creativity
    )
    result=response.choices[0].message['content']
    return result
```

```python
def tableOfContents(final_document):
    messages = [] 
    prompt_template=f"""Given below is a overall text containing various headings and description. Generate proper table of contents for only main headings.
    Output format:
    Contents:
    ---content---
    \nHere is the data:\n{final_document}
    """
    messages.append({'role':'user','content':prompt_template})
    response1 = openai.ChatCompletion.create(
        engine="neww",
        messages = messages,
        temperature=0,  # Adjust the temperature for creativitya
    )
    result=response1.choices[0].message['content']
    return result
```
